```python
from bs4 import BeautifulSoup
import urllib.request
import re
```

```python
url = "https://en.wikipedia.org/wiki/Artificial_intelligence"
page = urllib.request.urlopen(url)
soup = BeautifulSoup(page, 'html.parser')
```

```python
soup
```

## Find specific items in the page

```python
regex = re.compile('^tocsection-')
content_lis = soup.find_all('li', attrs={'class': regex})
content_lis
```

```python
content = []
for li in content_lis:
    content.append(li.getText().split('\n')[0])
content
```

```python
see_also_soup = soup.find('div', attrs={'class': 'div-col columns column-width'}).find_all('li')
see_also_soup
```

```python
see_also = []
for li in see_also_soup:
    a_tag = li.find('a',href=True, attrs={'title':True, 'class':False})
    href = a_tag['href']
    text = a_tag.getText()
    see_also.append([href, text])
see_also
```

## Saving data

```python
with open('content.txt', 'w') as f:
    for i in content:
        f.write(i+"\n")
```

```python
with open('see_also.csv', 'w') as f:
    for i in see_also:
        f.write(",".join(i)+"\n")
```
