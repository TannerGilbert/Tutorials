## Download images

```python
for file, folder in [('Cats.csv', 'Cats'), ('Cows.csv', 'Cows'), 
                     ('Dogs.csv', 'Dogs'), ('Horses.csv', 'Horses')]:
    dest = path/folder # path + '/' + folder
    dest.mkdir(parents=True, exist_ok=True)
    download_images(path/file, dest, max_pics=200)
```

## Verify images

```python
doc(verify_images)
```

```python
for folder in ('Cats', 'Cows', 'Dogs', 'Horses'):
    print(folder)
    verify_images(path/folder, delete=True, max_size=500)
```

## View data

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train='.', valid_pct=0.2,
                                  ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
data.classes
```

```python
doc(data.show_batch)
```

```python
data.show_batch(rows=3, figsize=(7, 8))
```

```python
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

## Initial training

```python
from fastai.metrics import error_rate # 1 - accuracy
learn = create_cnn(data, models.resnet34, metrics=error_rate)
```

```python
learn
```

```python
doc(learn.fit_one_cycle)
```

```python
defaults.device = torch.device('cuda') # makes sure the gpu is used
learn.fit_one_cycle(4)
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(4, max_lr=slice(3e-5, 3e-4))
```

```python
learn.save('animal-detection-stage-1')
```

## Cleaning data

```python
from fastai.widgets import *
```

```python
doc(DatasetFormatter().from_toplosses)
```

```python
ds, idxs = DatasetFormatter().from_toplosses(learn)
```

```python
ImageCleaner(ds, idxs, path)
```

## Training model

```python
df = pd.read_csv(path/'cleaned.csv', header='infer')
df.head()
```

```python
df[(df['name'].apply(lambda x: len(x)<5))]
```

```python
np.random.seed(42)
db = (ImageItemList.from_df(df, path)
                   .random_split_by_pct(0.2)
                   .label_from_df()
                   .transform(get_transforms(), size=224)
                   .databunch(bs=8)).normalize(imagenet_stats)
```

```python
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

```python
db.classes, db.c, len(db.train_ds), len(db.valid_ds)
```

```python
db.show_batch()
```

```python
learn.data = db
```

```python
learn.freeze()
learn.fit_one_cycle(4)
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(4, max_lr=slice(3e-5, 3e-4))
```

```python
learn.save('animal-detection-stage-2')
```

## Interpretation

```python
interp = ClassificationInterpretation.from_learner(learn)
```
